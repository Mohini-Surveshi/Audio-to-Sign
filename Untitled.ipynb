```python
from ibm_watson import SpeechToTextV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

Setup STT Service

```python
apikey = 'F_obQR5Y_RlmtF4Bk7gH7aDgqHUp5c2heyp-_HEyHwOZ'
url = 'https://api.eu-gb.speech-to-text.watson.cloud.ibm.com/instances/fe40d239-8c1b-4c60-9b9c-841974576cf2'
```

```python
#setup Service
authenticator= IAMAuthenticator(apikey)
stt= SpeechToTextV1(authenticator=authenticator)
stt.set_service_url(url)
```

Open Audio Source and Convert

```python
# Perform conversion
with open('Untitled.mp3', 'rb') as f:
    res = stt.recognize(audio=f, content_type='audio/mp3', model='en-US_NarrowbandModel', continuous=True).get_result()
```

```python
text = res['results'][0]['alternatives'][0]['transcript']
text
```
